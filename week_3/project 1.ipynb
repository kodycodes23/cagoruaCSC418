```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from IPython.display import display, Image

def user_authentication():
    email = input("Enter your email: ")
    age = int(input("Enter your age: "))
    if age < 18:
        print("Sorry, you must be above 18 years old to access the application.")
        return False, None
    return True, email

def select_art_category():
    print("Select an art category:")
    print("1. Traditional Art")
    print("2. Modern Art")
    print("3. Contemporary Art")
    choice = int(input("Enter your choice (1/2/3): "))
    categories = ["Traditional Art", "Modern Art", "Contemporary Art"]
    return categories[choice - 1]

def transform_image(img_path, transformation_matrix):
    img = cv2.imread(img_path)
    transformed_img = cv2.warpPerspective(img, transformation_matrix, (img.shape[1], img.shape[0]))
    return transformed_img

def display_transformed_image(img_path, transformation_matrix):
    transformed_img = transform_image(img_path, transformation_matrix)
    plt.imshow(cv2.cvtColor(transformed_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Example transformation matrix (for demonstration purposes)
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])

# Sample image paths for each art category
art_category_images = {
    "Traditional Art": "img/trad.jpg",
    "Modern Art": "img/modern.jpg",
    "Contemporary Art": "img/contemporary.jpg"
}

# Main function
def main():
    authenticated, email = user_authentication()
    if authenticated:
        category = select_art_category()
        img_path = art_category_images[category]
        display_transformed_image(img_path, M)

# Run the main function
main()
```
